```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
%matplotlib inline
```

```python
inputData = pd.read_csv('CanadaPerCapitaIncome.csv')
inputData
```

```python
plt.xlabel('Year', color='yellow')
plt.ylabel('Income Per Capita in USD', color='yellow')
plt.scatter(inputData.year, inputData['per capita income (US$)'], edgecolors='yellow',
color='yellow')
```

```python
regressor = linear_model.LinearRegression().fit(inputData[['year']], inputData['per capita income '
                                                                          '(US$)'])
```

```python
regressor
```

```python
regressor.predict([[2017], [2018], [2019], [2020]])
```
