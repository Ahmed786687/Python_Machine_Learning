```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline
from sklearn import linear_model
```

```python
data = pd.read_csv('./data.csv')
```

```python
data
```

```python
%matplotlib inline
```

```python
plt.xlabel('Area')
plt.ylabel('Price (US $)')
plt.scatter(data['area'], data['price'],)
```

```python
linearRegressor = linear_model.LinearRegression();
linearRegressor.fit(data[['area']], data.price)
```

### Here we can pass 'AREA' in predict() then it will give us "PRICE" for that area

```python
linearRegressor.predict([[3300], [5000]])
```

```python
plt.xlabel('Area')
plt.ylabel('Price (US $)')
plt.scatter(data['area'], data['price'],)
plt.plot(data.area, linearRegressor.predict(data[['area']]), color='blue')
```

#### Let's find HOW DOES THE REGRESSOR PREDICT THE VALUES

```python
linearRegressor.coef_
```

```python
linearRegressor.intercept_
```
