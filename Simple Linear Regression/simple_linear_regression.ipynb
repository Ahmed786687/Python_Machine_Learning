# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
dataset
```

```python
x_axis = dataset.iloc[:, :-1]
x_axis
```

```python
y_axis = dataset.iloc[:,-1]
y_axis
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_axis_train, x_axis_test, y_axis_train, y_axis_test = train_test_split(x_axis, y_axis,
                                                                        test_size=0.2,
                                                                        random_state=0)
```

```python
y_axis_train
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_axis_train, y_axis_train)
```

## Predicting the Test set results

```python
predictedSalary = regressor.predict(x_axis_train)
```

```python
predictedSalary
```

## Visualising the Training set results

```python
plt.scatter(x_axis_train, y_axis_train,color='red')
plt.plot(x_axis_train, predictedSalary, color="blue")
plt.title('Salary Vs. Experience (Training Dataset)')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(x_axis_test, y_axis_test,color='red')
plt.plot(x_axis_train, predictedSalary, color="blue") # it will remain same for test set
plt.title('Salary Vs. Experience (Training Dataset)')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
